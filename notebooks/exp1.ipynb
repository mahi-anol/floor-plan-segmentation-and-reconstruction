```python
import xml.etree.ElementTree as ET
```

```python
# Use a set to collect unique tags
unique_tags = set()
def findout_unqiue_xml_tags(file_path):
    # Load your SVG file
    tree = ET.parse(file_path)
    root = tree.getroot()

    for elem in root.iter():
        tag = elem.tag.split('}')[-1]
        unique_tags.add(tag)

unique_classes = set()
def findout_unique_class_names(file_path):
    # Load SVG file
    with open(file_path) as f:
        svg = f.read()

    # Parse SVG
    soup = BeautifulSoup(svg, "xml")

    # Collect all class attributes
    for tag in soup.find_all(class_=True):  # Finds all tags with a class attribute
        for c in tag["class"].split():
            unique_classes.add(c)

unique_texts = set()
def extract_unique_texts(file_path):
    # Load SVG file
    with open(file_path, encoding="utf-8") as f:
        svg = f.read()

    # Parse SVG
    soup = BeautifulSoup(svg, "xml")

    # Extract text content
    texts = [t.get_text(strip=True) for t in soup.find_all("text")]

    # Add each text string to the set
    for txt in texts:
        unique_texts.add(txt)
```
